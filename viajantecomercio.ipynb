```python
import numpy as np
import matplotlib.pyplot as plt
from IPython import display as ipythondisplay
import time

# Generar 20 ciudades aleatorias
n_cities = 20
cities = np.random.rand(n_cities, 2)

def draw_tsp(visited_indices, current_city_idx=None):
    plt.figure(figsize=(8, 6))
    plt.scatter(cities[:, 0], cities[:, 1], c='red', s=50, zorder=5)

    # Dibujar camino recorrido
    if len(visited_indices) > 1:
        path = cities[visited_indices]
        plt.plot(path[:, 0], path[:, 1], 'skyblue', linewidth=2, alpha=0.8)

    # Resaltar ciudad actual
    if current_city_idx is not None:
        plt.scatter(cities[current_city_idx, 0], cities[current_city_idx, 1], c='lime', s=150, edgecolors='black')

    plt.title("Heurística: Vecino más cercano (TSP)")
    plt.xlim(-0.1, 1.1); plt.ylim(-0.1, 1.1)

    ipythondisplay.clear_output(wait=True)
    ipythondisplay.display(plt.gcf())
    plt.close()
    time.sleep(0.3)

# Algoritmo con Heurística Greedy
unvisited = list(range(n_cities))
current = unvisited.pop(0)
tour = [current]

while unvisited:
    draw_tsp(tour, current)
    # Heurística: Elegir el que tenga menor distancia euclidiana
    current = min(unvisited, key=lambda city: np.linalg.norm(cities[current] - cities[city]))
    unvisited.remove(current)
    tour.append(current)

tour.append(tour[0]) # Volver al inicio
draw_tsp(tour)
```
