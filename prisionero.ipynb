```python
import numpy as np
import matplotlib.pyplot as plt
from IPython import display as ipythondisplay
import time

# 0: Cooperar (Callar), 1: Traicionar (Delatar)
payoff_matrix = {
    (0, 0): (3, 3),   # Ambos callan: +3 puntos
    (0, 1): (0, 5),   # A calla, B delata: A pierde, B gana mucho
    (1, 0): (5, 0),   # A delata, B calla: A gana mucho, B pierde
    (1, 1): (1, 1)    # Ambos delatan: +1 punto (mal para ambos)
}

def simulate_prison():
    history_a, history_b = [], []
    score_a, score_b = 0, 0
    scores_over_time = []

    fig, ax = plt.subplots(figsize=(8, 4))

    for i in range(30):
        # Estrategia: A veces cooperan, a veces traicionan (aleatorio para el ejemplo)
        choice_a = np.random.choice([0, 1], p=[0.7, 0.3])
        choice_b = np.random.choice([0, 1], p=[0.6, 0.4])

        pa, pb = payoff_matrix[(choice_a, choice_b)]
        score_a += pa
        score_b += pb
        scores_over_time.append((score_a, score_b))

        ax.clear()
        data = np.array(scores_over_time)
        ax.plot(data[:, 0], label="Jugador A (Cooperador)", color="teal", linewidth=2)
        ax.plot(data[:, 1], label="Jugador B (Agresivo)", color="tomato", linewidth=2)
        ax.set_title(f"Dilema del Prisionero Iterado - Ronda {i+1}")
        ax.legend()
        ax.grid(True, alpha=0.3)

        ipythondisplay.clear_output(wait=True)
        ipythondisplay.display(fig)
        time.sleep(0.2)
    plt.close()

simulate_prison()
```
